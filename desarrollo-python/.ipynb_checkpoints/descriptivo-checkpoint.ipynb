# Modelación

```python
import pickle

with open("models/scaler_model.pkl", 'rb') as file:
    scaler = pickle.load(file)

with open("models/grid_search_lgb.pkl", 'rb') as file:
    grid_search_lgb = pickle.load(file)

with open("models/grid_search_rfc.pkl", 'rb') as file:
    grid_search_rfc = pickle.load(file)

with open("models/grid_search_xgb.pkl", 'rb') as file:
    grid_search_xgb = pickle.load(file)

with open("models/X_test.pkl", 'rb') as file:
    X_test = pickle.load(file)

with open("models/y_test.pkl", 'rb') as file:
    y_test = pickle.load(file)
```

En el archivo modelado.py se encuentra el tratamiento a las variables categóricas y numéricas (Feature Engineer) y el benchmark de los modelos utilizados. Para este ejercidos de clasificación opte por modelos basados en ensambles, puntualmente el GradientBoostingClassifier, RandomForestClassifier y el LGBMClassifier.

Se realiza su respectivos tuning de parámetros y la validación cruzada con un k=4
Se cargan los pkl que contiene las clases con los modelos y las clases con los ajustes del procesamiento de los datos.


```python
resultado = pd.read_excel(r"resultado_metricas.xlsx")
resultado
```

La tabla anterior presenta los resultados de los modelos comparados, en términos general sus métricas tanto en entrenamiento como en test se parece mucho, para este analizaremos las métricas del test, donde los promedio del accuracy están alrededor del 99%, la presión promedio de los modelos fue 99%. 

En terminos general los 3 modelos tiene rendimientos muy buenos, el GradientBoostingClassifier y RandomForestClassifier sus ajustes son perfectos para este ejerccio se seleccionara el modelo seleccionado es el RandomForestClassifier

```python
y_test_pred = pd.DataFrame(grid_search_rfc.predict(X_test))
predicciones_probabilidades = grid_search_rfc.predict_proba(X_test)

modelado.grafica_curva_auc(clases_reales=y_test, predicciones=y_test_pred,
                           predicciones_probabilidades=predicciones_probabilidades)
```
